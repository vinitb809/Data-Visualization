```python
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd

df = pd.read_csv('C://Users//VINIT//Downloads//Work/DMV//Assignment//120-years-of-olympic-history-athletes-and-results//taableau.csv')
df['count'] = df.groupby('Team')['Team'].transform('count')
newdf = df.drop_duplicates(subset='NOC')

fig = go.Figure(data=go.Choropleth(
    locations = newdf['NOC'],
    z = newdf['count'],
    text = newdf['Team'],
    colorscale='greens',
    autocolorscale=False,
    colorbar_title= "Medal Counts"
))
fig.update_layout(
    title_text='Medals per country',
    geo=dict(
        showframe=False,
        showcoastlines=False,
        projection_type='equirectangular'
    ),
        
    
)

fig.show()

```
